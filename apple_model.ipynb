```python
import pandas as pd

# ---------- Income Statement ----------
is_clean = is_df[[
    "endDate",
    "totalRevenue",                 # Revenue
    "costOfRevenue",                # COGS
    "researchDevelopment",          # R&D
    "sellingGeneralAdministrative", # SG&A
    "interestExpense",              # Interest
    "incomeTaxExpense",             # Taxes
    "netIncome"                     # Net Income
]].copy()

# ---------- Balance Sheet ----------
bs_clean = bs_df[[
    "endDate",
    "cash",
    "netReceivables",               # AR
    "inventory",
    "accountsPayable",              # AP
    "propertyPlantEquipment",       # PP&E (gross or net as provided)
    "shortLongTermDebt",            # Short-term debt/current portion
    "longTermDebt",                 # Long-term debt
    "totalAssets",
    "totalLiab",
    "totalStockholderEquity"
]].copy()

# ---------- Cash Flow ----------
# Note: capex & dividends are usually negative in CF; make them positive outflows
cf_tmp = cf_df[[
    "endDate",
    "capitalExpenditures",
    "dividendsPaid",
    "depreciation",
    "totalCashFromOperatingActivities"
]].copy()

cf_clean = cf_tmp.rename(columns={
    "capitalExpenditures": "capex_raw",
    "dividendsPaid": "dividends_raw"
})
cf_clean["CapEx"] = -cf_clean["capex_raw"]              # positive number for spend
cf_clean["Dividends"] = -cf_clean["dividends_raw"]      # positive number for cash paid
cf_clean = cf_clean[["endDate", "CapEx", "Dividends", "depreciation", "totalCashFromOperatingActivities"]]

# ---------- Common formatting ----------
for df in (is_clean, bs_clean, cf_clean):
    df["endDate"] = pd.to_datetime(df["endDate"])
    # Convert to USD millions
    for c in df.columns:
        if c != "endDate":
            df[c] = df[c] / 1_000_000
    df.sort_values("endDate", inplace=True)
    df["Year"] = df["endDate"].dt.year

# ---------- Preview ----------
print("Income Statement (millions):")
display(is_clean)

print("Balance Sheet (millions):")
display(bs_clean)

print("Cash Flow (millions):")
display(cf_clean)
```

```python
# ---- Base Case Driver Seeding ----

hist = {}

# Revenue CAGR (2016–2019)
rev_start = is_clean.loc[is_clean["Year"] == 2016, "totalRevenue"].values[0]
rev_end   = is_clean.loc[is_clean["Year"] == 2019, "totalRevenue"].values[0]
cagr = (rev_end / rev_start) ** (1/3) - 1
hist["Revenue CAGR"] = cagr

# Avg margins
hist["COGS_pct"] = (is_clean["costOfRevenue"] / is_clean["totalRevenue"]).mean()
hist["R&D_pct"]  = (is_clean["researchDevelopment"] / is_clean["totalRevenue"]).mean()
hist["SG&A_pct"] = (is_clean["sellingGeneralAdministrative"] / is_clean["totalRevenue"]).mean()

# Effective tax rate (tax / pre-tax income)
hist["Tax_rate"] = (is_clean["incomeTaxExpense"] / 
                    (is_clean["netIncome"] + is_clean["incomeTaxExpense"])).mean()

# CapEx % Revenue
hist["CapEx_pct"] = (cf_clean["CapEx"] / is_clean["totalRevenue"]).mean()

# Dep % Revenue
hist["Dep_pct"] = (cf_clean["depreciation"] / is_clean["totalRevenue"]).mean()

# Dividend payout % (Div / NI)
hist["Div_payout"] = (cf_clean["Dividends"] / is_clean["netIncome"]).mean()

# Working capital days
rev = is_clean.set_index("Year")["totalRevenue"]
cogs = is_clean.set_index("Year")["costOfRevenue"]

# Averages across years
dso = (bs_clean["netReceivables"] / (rev / 365)).mean()
dsi = (bs_clean["inventory"] / (cogs / 365)).mean()
dpo = (bs_clean["accountsPayable"] / (cogs / 365)).mean()

hist["DSO"] = dso
hist["DSI"] = dsi
hist["DPO"] = dpo

# Interest rates (rough: Interest Expense / Avg Debt)
avg_debt = (bs_clean["shortLongTermDebt"] + bs_clean["longTermDebt"]).mean()
avg_int_exp = is_clean["interestExpense"].mean()
hist["Debt_rate"] = abs(avg_int_exp / avg_debt)  # take abs since interestExp is negative

# Show results nicely
pd.Series(hist).apply(lambda x: round(x,4) if isinstance(x, (float,int)) else x)
```

```python
# Base case assumptions derived from historicals
base_assumptions = {
    "Rev_Growth": [0.065, 0.05, 0.045, 0.04, 0.035],  # taper down CAGR
    "COGS_pct": [0.616]*5,
    "R&D_pct": [0.053]*5,
    "SG&A_pct": [0.066]*5,
    "Tax_rate": [0.21]*5,
    "CapEx_pct": [0.051]*5,
    "Dep_pct": [0.046]*5,
    "Div_payout": [0.25]*5,
    "ST_Debt_Rate": [0.027]*5,
    "LT_Debt_Rate": [0.027]*5,
    "DSO": [59.5]*5,
    "DSI": [9.1]*5,
    "DPO": [111.8]*5
}

years = list(range(2020, 2025))
assumptions_df = pd.DataFrame(base_assumptions, index=years)
assumptions_df.index.name = "Year"

assumptions_df
```

```python
# ---- Simple Base Case assumptions ----
base_df = assumptions_df.copy()

# ---- Bull case adjustments ----
# +3% growth, lower COGS, lower SG&A, lower Tax, better WC, higher CapEx
bull_df = base_df.copy()
bull_df["Rev_Growth"] = bull_df["Rev_Growth"] + 0.03
bull_df["COGS_pct"]   = bull_df["COGS_pct"] - 0.02
bull_df["SG&A_pct"]   = bull_df["SG&A_pct"] - 0.01
bull_df["Tax_rate"]   = bull_df["Tax_rate"] - 0.01
bull_df["CapEx_pct"]  = bull_df["CapEx_pct"] + 0.01
bull_df["DSO"]        = bull_df["DSO"] - 2
bull_df["DSI"]        = bull_df["DSI"] - 2
bull_df["DPO"]        = bull_df["DPO"] + 2

# ---- Bear case adjustments ----
# -3% growth, higher COGS, higher SG&A, higher Tax, worse WC, lower CapEx
bear_df = base_df.copy()
bear_df["Rev_Growth"] = bear_df["Rev_Growth"] - 0.03
bear_df["COGS_pct"]   = bear_df["COGS_pct"] + 0.02
bear_df["SG&A_pct"]   = bear_df["SG&A_pct"] + 0.01
bear_df["Tax_rate"]   = bear_df["Tax_rate"] + 0.01
bear_df["CapEx_pct"]  = bear_df["CapEx_pct"] - 0.01
bear_df["DSO"]        = bear_df["DSO"] + 2
bear_df["DSI"]        = bear_df["DSI"] + 2
bear_df["DPO"]        = bear_df["DPO"] - 2

# Combine into one DataFrame
assumptions_all = pd.concat({
    "Base": base_df,
    "Bull": bull_df,
    "Bear": bear_df
}, names=["Scenario", "Year"])

assumptions_all
```

```python
# ---- Income Statement Projection ----

def project_income_statement(hist_df, assumptions, scenario="Base"):
    """
    Projects 5-year Income Statement based on assumptions for a given scenario.
    """
    # Start with last actual revenue (2019)
    base_rev = hist_df.loc[hist_df["Year"] == 2019, "totalRevenue"].values[0]
    
    # Extract scenario assumptions
    scen = assumptions.loc[scenario]
    
    proj_rows = []
    rev = base_rev
    
    for yr, row in scen.iterrows():
        # Revenue grows from prior year
        rev = rev * (1 + row["Rev_Growth"])
        cogs = rev * row["COGS_pct"]
        rnd  = rev * row["R&D_pct"]
        sga  = rev * row["SG&A_pct"]
        dep  = rev * row["Dep_pct"]   # simple % of revenue
        ebit = rev - cogs - rnd - sga - dep
        int_exp = -( (row["ST_Debt_Rate"] + row["LT_Debt_Rate"]) / 2 * 50_000 )  # placeholder debt
        ebt  = ebit + int_exp
        tax  = max(ebt,0) * row["Tax_rate"]
        ni   = ebt - tax
        
        proj_rows.append({
            "Year": yr,
            "Revenue": rev,
            "COGS": cogs,
            "R&D": rnd,
            "SG&A": sga,
            "Dep": dep,
            "EBIT": ebit,
            "Interest": int_exp,
            "EBT": ebt,
            "Tax": tax,
            "Net_Income": ni
        })
    
    return pd.DataFrame(proj_rows)

# Run for Base scenario
is_proj_base = project_income_statement(is_clean, assumptions_all, scenario="Base")
is_proj_base
```

```python
# Run projections for all 3 scenarios
proj_all = []

for scen in ["Base", "Bull", "Bear"]:
    df = project_income_statement(is_clean, assumptions_all, scenario=scen)
    df["Scenario"] = scen
    proj_all.append(df)

is_proj_all = pd.concat(proj_all).set_index(["Scenario", "Year"])
is_proj_all
```

```python
# ---- Step 12: Balance Sheet Projection (simple; Cash as plug) ----

# Grab 2019 actual anchors (millions)
bs_2019 = bs_clean.sort_values("Year").iloc[-1]
cash0   = bs_2019["cash"]
ar0     = bs_2019["netReceivables"]
inv0    = bs_2019["inventory"]
ap0     = bs_2019["accountsPayable"]
ppe0    = bs_2019["propertyPlantEquipment"]
stdebt0 = bs_2019["shortLongTermDebt"]
ltdebt0 = bs_2019["longTermDebt"]
assets0 = bs_2019["totalAssets"]
liab0   = bs_2019["totalLiab"]
equity0 = bs_2019["totalStockholderEquity"]

# Residual buckets we will keep constant for simplicity
other_assets0 = assets0 - (cash0 + ar0 + inv0 + ppe0)
other_liab0   = liab0   - (ap0   + stdebt0 + ltdebt0)

def project_balance_sheet_simple(is_proj_df, assumptions_all, scenario="Base"):
    """
    Simple BS projection using:
      - AR/Inv/AP from DSO/DSI/DPO
      - PP&E roll-forward from CapEx % Rev and Dep from IS projection
      - Equity roll-forward from NI - Dividends
      - ST/LT debt held flat at 2019 levels
      - Cash as BALANCING ITEM (temporary)
    """
    # 2019 anchors
    prev_ar, prev_inv, prev_ap = ar0, inv0, ap0
    prev_ppe = ppe0
    prev_equity = equity0
    prev_cash = cash0

    rows = []
    # Scenario slice
    scen = assumptions_all.loc[scenario].copy()
    # Ensure year order
    scen = scen.sort_index()

    for yr, a in scen.iterrows():
        # Pull IS drivers for this year
        row_is = is_proj_df.loc[is_proj_df["Year"] == yr].iloc[0]
        rev  = row_is["Revenue"]
        cogs = row_is["COGS"]
        dep  = row_is["Dep"]
        ni   = row_is["Net_Income"]

        # Working capital from days
        ar  = (rev/365.0) * a["DSO"]
        inv = (cogs/365.0) * a["DSI"]
        ap  = (cogs/365.0) * a["DPO"]

        # PP&E roll-forward
        capex = rev * a["CapEx_pct"]
        ppe   = prev_ppe + capex - dep

        # Equity roll-forward (assume other equity items flat; change via retained earnings only)
        div   = a["Div_payout"] * ni
        equity = prev_equity + (ni - div)

        # Hold debts flat (we'll introduce ST-debt plug in the next step)
        stdebt = stdebt0
        ltdebt = ltdebt0

        # Cash as residual to satisfy A = L + E
        # A = Cash + AR + Inv + PPE + OtherAssets
        # L+E (ex Cash) = AP + STDebt + LTDebt + OtherLiab + Equity
        cash = (ap + stdebt + ltdebt + other_liab0 + equity) - (ar + inv + ppe + other_assets0)

        # Save row
        rows.append({
            "Year": yr,
            "Revenue": rev,
            "COGS": cogs,
            "Dep": dep,
            "Net_Income": ni,
            "CapEx": capex,
            "Dividends": div,
            "AR": ar,
            "Inventory": inv,
            "AP": ap,
            "PP&E": ppe,
            "Cash": cash,
            "ST_Debt": stdebt,
            "LT_Debt": ltdebt,
            "Other_Assets": other_assets0,
            "Other_Liab": other_liab0,
            "Equity": equity,
            "Total_Assets": cash + ar + inv + ppe + other_assets0,
            "Total_Liab_Equity": ap + stdebt + ltdebt + other_liab0 + equity
        })

        # Update prevs for next year
        prev_ar, prev_inv, prev_ap = ar, inv, ap
        prev_ppe = ppe
        prev_equity = equity
        prev_cash = cash

    bs_proj = pd.DataFrame(rows).set_index("Year")
    return bs_proj

# Run for Base scenario using the IS projection we already computed
bs_proj_base = project_balance_sheet_simple(is_proj_base, assumptions_all, scenario="Base")

# Quick check: A = L + E (should be ~equal; small rounding noise OK)
bs_proj_base[["Total_Assets", "Total_Liab_Equity", "Cash", "AR", "Inventory", "AP", "PP&E", "Equity"]]
```

```python
# ---- Step 13: Cash Flow Projection (indirect method) ----

def project_cashflow(is_proj_df, bs_proj_df):
    """
    Build Cash Flow from NI, Dep, ΔWC, CapEx, Dividends.
    """
    rows = []
    prev_ar = bs_proj_df.iloc[0]["AR"]
    prev_inv = bs_proj_df.iloc[0]["Inventory"]
    prev_ap = bs_proj_df.iloc[0]["AP"]
    prev_cash = bs_proj_df.iloc[0]["Cash"]

    for yr in bs_proj_df.index:
        ni   = is_proj_df.loc[is_proj_df["Year"] == yr, "Net_Income"].values[0]
        dep  = is_proj_df.loc[is_proj_df["Year"] == yr, "Dep"].values[0]
        capex = bs_proj_df.loc[yr, "CapEx"]
        div   = bs_proj_df.loc[yr, "Dividends"]

        ar = bs_proj_df.loc[yr, "AR"]
        inv = bs_proj_df.loc[yr, "Inventory"]
        ap = bs_proj_df.loc[yr, "AP"]

        # ΔWorking Capital = ΔAR + ΔInv − ΔAP
        if yr != bs_proj_df.index[0]:
            delta_wc = (ar - prev_ar) + (inv - prev_inv) - (ap - prev_ap)
        else:
            delta_wc = 0

        CFO = ni + dep - delta_wc
        CFI = -capex
        CFF = -div  # no debt change for now
        net_cf = CFO + CFI + CFF
        end_cash = prev_cash + net_cf if yr != bs_proj_df.index[0] else bs_proj_df.loc[yr, "Cash"]

        rows.append({
            "Year": yr,
            "CFO": CFO,
            "CFI": CFI,
            "CFF": CFF,
            "Net_CF": net_cf,
            "End_Cash": end_cash
        })

        prev_ar, prev_inv, prev_ap = ar, inv, ap
        prev_cash = end_cash

    return pd.DataFrame(rows).set_index("Year")

# Build CF for Base scenario
cf_proj_base = project_cashflow(is_proj_base, bs_proj_base)
cf_proj_base
```

```python
# ---- Full 3-Statement Projection Engine ----

def run_projection(hist_is, hist_bs, assumptions_all, scenario="Base"):
    # Income Statement
    is_proj = project_income_statement(hist_is, assumptions_all, scenario=scenario)
    # Balance Sheet
    bs_proj = project_balance_sheet_simple(is_proj, assumptions_all, scenario=scenario)
    # Cash Flow
    cf_proj = project_cashflow(is_proj, bs_proj)
    return is_proj, bs_proj, cf_proj

# Run for all scenarios
results = {}
for scen in ["Base", "Bull", "Bear"]:
    is_proj, bs_proj, cf_proj = run_projection(is_clean, bs_clean, assumptions_all, scenario=scen)
    results[scen] = {"IS": is_proj, "BS": bs_proj, "CF": cf_proj}

# Quick peek at Net Income across scenarios
pd.concat({s: results[s]["IS"][["Net_Income"]] for s in results}, axis=1)
```

```python
# ---- KPI / Ratios Builder ----
def build_kpis(is_proj, bs_proj, cf_proj, scen):
    df = pd.DataFrame(index=is_proj["Year"])
    
    rev = is_proj.set_index("Year")["Revenue"]
    cogs = is_proj.set_index("Year")["COGS"]
    ebit = is_proj.set_index("Year")["EBIT"]
    ni   = is_proj.set_index("Year")["Net_Income"]
    
    capex = bs_proj["CapEx"]
    dep   = is_proj.set_index("Year")["Dep"]
    cfo   = cf_proj["CFO"]
    
    # Gross / EBIT / Net Margins
    df["GrossMargin_pct"] = (rev - cogs) / rev
    df["EBITMargin_pct"]  = ebit / rev
    df["NetMargin_pct"]   = ni / rev
    
    # Free Cash Flow
    df["FCF"] = cfo - capex
    
    # ROIC (proxy)
    debt = bs_proj["ST_Debt"] + bs_proj["LT_Debt"]
    equity = bs_proj["Equity"]
    df["ROIC_pct"] = ni / (equity + debt)
    
    # Leverage ratio
    df["Leverage"] = debt / equity
    
    # Working capital days → CCC
    df["DSO"] = assumptions_all.loc[scen]["DSO"].values
    df["DSI"] = assumptions_all.loc[scen]["DSI"].values
    df["DPO"] = assumptions_all.loc[scen]["DPO"].values
    df["CCC"] = df["DSO"] + df["DSI"] - df["DPO"]
    
    df["Scenario"] = scen
    return df

# Build KPIs for all scenarios
kpi_all = pd.concat([build_kpis(results[s]["IS"], results[s]["BS"], results[s]["CF"], s) 
                     for s in results])
kpi_all.set_index(["Scenario"], append=True)
```

```python
# Create exports folder
import os
EXPORT_DIR = "exports"
os.makedirs(EXPORT_DIR, exist_ok=True)

# ---- Flatten & export IS/BS/CF for all scenarios ----
import pandas as pd

def flatten_outputs(results, key):
    frames = []
    for scen, d in results.items():
        df = d[key].copy()
        if "Year" in df.columns:
            pass
        else:
            df = df.reset_index()  # BS/CF might be indexed by Year
        df["Scenario"] = scen
        frames.append(df)
    out = pd.concat(frames, ignore_index=True)
    # Put columns in a nice order
    cols = ["Scenario", "Year"] + [c for c in out.columns if c not in ("Scenario","Year")]
    return out[cols]

is_out = flatten_outputs(results, "IS")
bs_out = flatten_outputs(results, "BS")
cf_out = flatten_outputs(results, "CF")

# ---- KPIs table (we already created kpi_all) ----
kpi_out = kpi_all.copy().reset_index()  # index is Year; Scenario stored as column in build_kpis
# Ensure columns are ordered
if "Scenario" in kpi_out.columns and "Year" in kpi_out.columns:
    kpi_out = kpi_out[["Scenario","Year"] + [c for c in kpi_out.columns if c not in ("Scenario","Year")]]

# ---- Quick sanity checks before export ----
# Balance sheet should balance
bs_out["Balance_Check"] = (bs_out["Total_Assets"] - bs_out["Total_Liab_Equity"]).round(3)
# Cash tie (End_Cash vs BS Cash) per scenario/year
cf_vs_bs = cf_out.merge(bs_out[["Scenario","Year","Cash"]], on=["Scenario","Year"], how="left")
cf_vs_bs["Cash_Tie"] = (cf_vs_bs["End_Cash"] - cf_vs_bs["Cash"]).round(3)

print("Max BS imbalance:", bs_out["Balance_Check"].abs().max())
print("Max Cash tie delta:", cf_vs_bs["Cash_Tie"].abs().max())

# ---- Write CSVs ----
is_path  = os.path.join(EXPORT_DIR, "is_all.csv")
bs_path  = os.path.join(EXPORT_DIR, "bs_all.csv")
cf_path  = os.path.join(EXPORT_DIR, "cf_all.csv")
kpi_path = os.path.join(EXPORT_DIR, "kpi_all.csv")

is_out.to_csv(is_path, index=False)
bs_out.to_csv(bs_path, index=False)
cf_out.to_csv(cf_path, index=False)
kpi_out.to_csv(kpi_path, index=False)

is_path, bs_path, cf_path, kpi_path
```
